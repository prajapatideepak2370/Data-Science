# NAIVE BAYE'S  <br>
Classification
<br>
P(B/A) = P(B) * P(A/B) / P(A)

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Sample dataset
data = {
    "Age": ["Young", "Young", "Middle", "Old", "Old", "Old", "Middle", "Young", "Young", "Old"],
    "Income": ["High", "High", "High", "Medium", "Low", "Low", "Low", "Medium", "Low", "Medium"],
    "Buy": ["No", "No", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes", "Yes"]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Convert Text into Numbers

le_age = LabelEncoder()
le_income = LabelEncoder()
le_buy = LabelEncoder()

df["Age"] = le_age.fit_transform(df["Age"])
df["Income"] = le_income.fit_transform(df["Income"])
df["Buy"] = le_buy.fit_transform(df["Buy"])

print(df)
```

Age:   <br>
Middle = 0   <br>
Old = 1   <br>
Young = 2   <br>
   <br>
Income:   <br>
High = 0   <br>
Low = 1   <br>
Medium = 2   <br>
   <br>
Buy:   <br>
No = 0   <br>
Yes = 1

```python
# Split Features and Target
X = df[['Age', 'Income']]
Y = df['Buy']
```

Train Naive Bayes Model

```python
from sklearn.naive_bayes import CategoricalNB

model = CategoricalNB()

model.fit(X, Y)
```

Make Prediction

Predict for:

Age = Young <br>
Income = Medium

```python
test = [[2, 2]]   # Young, Medium

prediction = model.predict(test)

if prediction[0] == 1:
    print("Prediction: Buy")
else:
    print("Prediction: Don't Buy")
```

Naive Bayes is based on Bayes' Theorem:


Age = Young <br>
Income = Medium <br>

Dataset:<br>

No	Age	Income	Buy<br>
1	Young	High	No<br>
2	Young	High	No<br>
3	Middle	High	Yes<br>
4	Old	Medium	Yes<br>
5	Old	Low	Yes<br>
6	Old	Low	No<br>
7	Middle	Low	Yes<br>
8	Young	Medium	No<br>
9	Young	Low	Yes<br>
10	Old	Medium	Yes<br>
<br>
We want to predict:<br>
<br>
Buy = Yes ? or Buy = No ?<br>
<br>
Step 1: Bayes Theorem
P(A∣B)= P(B∣A)*P(A)/P(B)
<br>	​
<br>
P(A∣B)= 0.85(0.20)/0.25=0.68<br>
<br>
P(B)=0.25        <br>
P(B|A)P(A)=0.17       <br>
P(A|B)~0.68       <br>
Posterior = useful evidence / total evidence       <br>
       <br>
For comparison, we calculate:       <br>
       <br>
Score(Yes) = P(Yes) × P(Young|Yes) × P(Medium|Yes)       <br>
Score(No) = P(No) × P(Young|No) × P(Medium|No)       <br>
       <br>
The class with the larger score is the prediction.       <br>
       <br>
Step 2: Find Prior Probabilities       <br>
       <br>
Count the target values.       <br>
       <br>
Buy = Yes       <br>
       <br>
Rows: 3, 4, 5, 7, 9, 10  
       <br>
Total = 6       <br>
       <br> 
         ->      P(Yes)= 6/10 = 0.6      <br>
      <br>
Buy = No      <br>
      <br>
Rows: 1,2,6,8      <br>
    <br>
Total = 4    <br>
         ->      P(No)= 4/10 = 0.4     <br>
    <br>
Step 3: Calculate Likelihoods    <br>
    <br>
A. P(Young | Yes)    <br>
  ->   Among the 6 "Yes" rows:    <br>
       ->     Young appears 1 time.    <br>
		 ->             P(Young∣Yes)=61​=0.1667    <br>
    <br>
B. P(Medium | Yes)    <br>
  ->   Among the Yes rows    <br>
       ->     Medium appears 2 times.    <br>
            ->          P(Medium∣Yes)= 2/6 = 0.3333    <br>
    <br>
C. P(Young | No)    <br>
   ->    "No" rows    <br>
     ->          Young appears 3 times.    <br>
          ->             P(Young∣No)= 3/4 = 0.75    <br>
    <br>
D. P(Medium | No)    <br>
  ->   Income values for No:    <br>
 ->  	     Medium appears 1 time.    <br>
     ->                P(Medium∣No)= 1/4 = 0.25    <br>
    <br>

Step 4: Calculate Score for "Yes"    <br>
Score(Yes)=P(Yes)×P(Young∣Yes)×P(Medium∣Yes)    <br>
    <br>
Substitute the values:  0.6×0.1667×0.3333     <br>
    <br>
First,    <br>
--> 0.1667×0.3333=0.0556    <br>
    <br>
Now,    <br>
--> 0.6×0.0556=0.0333    <br>
    <br>
So,    <br>
-> Score(Yes)=0.0333    <br>
   <br>
Step 5: Calculate Score for "No"   <br>
   <br>
 ----> Score(No)=P(No)×P(Young∣No)×P(Medium∣No)   <br>
   <br>
Substitute values:  0.4×0.75×0.25    <br>
   <br>
First,   <br>
--> 0.75×0.25=0.1875   <br>
   <br>
Now,   <br>
--> 0.4×0.1875=0.075   <br>
   <br>
So,   <br>
-> Score(No)=0.075   <br>
	​   <br>
​Step 6: Compare the Scores	​   <br>
| Class | Score  |
| ----- | ------ |
| Yes   | 0.0333 |
| No    | 0.075  |
	​   <br>
Since	​   <br>
 -->    0.075 > 0.0333
	​   <br>
the larger score is No.	​   <br>
	​   <br>
Final Prediction
--> Buy = No
	​	​   <br>
This matches what the Naive Bayes classifier predicts for Age = Young and Income = Medium.	​   <br>
  <br>
