# UNIVARIATE ANALYSIS

Univariate Analysis is the process of analyzing a single variable at a time to understand its characteristics, distribution, and patterns

Objectives of Univariate Analysis
- Understand the distribution of the variable.
- Identify the central tendency (mean, median, mode).
- Measure the spread of the data (range, variance, standard deviation).
- Detect outliers and missing values.
- Understand the frequency or proportion of different categories.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('train.csv')
```

```python
dataset.sample(5)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

 1. Categorical Data



```python
dataset['Survived'].value_counts()
```

```python
# Set the style
sns.set_style("whitegrid")

# Create the countplot
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=dataset, palette='Set2')

# Add title and labels
plt.title('Count of Survived Passengers')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
sns.countplot(x='Pclass', data=dataset, palette='Set2')

plt.title('Passenger Class')
```

```python
dataset['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
plt.title('Sex Distribution Graph')
```

```python
sns.countplot(x='Embarked', data=dataset, palette='Set2')
plt.xlabel('Embarked (S = Southampton, C = Cherbourg , Q = Queenstown)')
plt.ylabel('Count')
```

2. Numerical Data

```python
sns.histplot(dataset['Age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution')
```

```python
# Create a box plot for the Fare column
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataset['Fare'], color='skyblue')

# Add title and labels
plt.title('Box Plot of Fare')
plt.xlim(-5, 100)
```
