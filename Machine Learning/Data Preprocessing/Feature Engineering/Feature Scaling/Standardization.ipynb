# FEATURE SCALING

Feature scaling is the process of changing the values of numerical features so they are on a similar scale, without changing the meaning of the data.

# Standardization
Standardization is a feature scaling technique that transforms numerical data so that it has:

- Mean = 0
- Standard Deviation = 1

This makes features with different ranges comparable without changing the relationships between the data points.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df = df.iloc[:,2:]
```

```python
df.sample(5)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.3, random_state=0)

X_train.shape, X_test.shape
```

Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)
# Transfrom train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

Effect of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color='red')
ax2.set_title("After Scaling")

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

#before scaling 
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

#after scaling
ax2.set_title('After Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

Comparison of Distribution

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

#before scaling 
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

#After scaling 
ax2.set_title('Age Distribution After Scaling')
sns.kdeplot(X_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

#before scaling 
ax1.set_title('EstimatedSalary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

#After scaling 
ax2.set_title('EstimatedSalary Distribution After Scaling')
sns.kdeplot(X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

Why Scaling is important? 

Feature scaling helps:
- Make machine learning models learn faster.
- Prevent features with large values from dominating features with small values.
- Improve the accuracy of algorithms that depend on distances or gradients.

Common Feature Scaling Methods
1. Min-Max Scaling (Normalization)
- Converts values to a range between 0 and 1.
- Formula: x′=​x−xmin​​/xmax​−xmin

2. Standardization (Z-score Scaling)
- Transforms data so it has a mean of 0 and a standard deviation of 1.
- Formula: x′=x−μ​/σ

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual score: ", accuracy_score(y_test, y_pred))
print("Scaled score: ", accuracy_score(y_test, y_pred_scaled))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
dt_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("Actual score: ", accuracy_score(y_test, y_pred))
print("Scaled score: ", accuracy_score(y_test, y_pred_scaled))
```

When is Feature Scaling Needed?

Feature scaling is recommended for algorithms such as:

- K-Nearest Neighbors (KNN)
- K-Means Clustering
- Support Vector Machines (SVM)
- Logistic Regression
- Linear Regression (especially with gradient descent)
- Neural Networks

It is usually not required for tree-based algorithms such as:

- Decision Trees
- Random Forests
- Gradient Boosting (XGBoost, LightGBM, CatBoost)
