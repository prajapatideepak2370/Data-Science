# One-Hot Encoding

## Definition

**One-Hot Encoding** is a technique used in machine learning to convert **categorical variables into numerical format** by creating a separate binary column for each unique category. Each category is represented by a value of **1** in its corresponding column and **0** in all other columns.

It is mainly used for **nominal categorical data**, where categories do not have any natural order or ranking.

---

## Example of One-Hot Encoding

Suppose we have a dataset containing a **Color** feature:

| Color |
| ----- |
| Red   |
| Blue  |
| Green |
| Red   |
| Blue  |

After applying One-Hot Encoding:

| Color_Red | Color_Blue | Color_Green |
| --------- | ---------- | ----------- |
| 1         | 0          | 0           |
| 0         | 1          | 0           |
| 0         | 0          | 1           |
| 1         | 0          | 0           |
| 0         | 1          | 0           |

Here:

* **Red** is represented as `[1, 0, 0]`
* **Blue** is represented as `[0, 1, 0]`
* **Green** is represented as `[0, 0, 1]`

Each category gets its own column, and the presence of that category is represented by **1**.

---

## Why Do We Use One-Hot Encoding?

Machine learning algorithms work with numerical data, not text values. For example, a model cannot directly understand:

| City    |
| ------- |
| Delhi   |
| Mumbai  |
| Chennai |

So we convert these categories into numbers.

However, using simple label encoding:

| City    | Encoded |
| ------- | ------- |
| Delhi   | 0       |
| Mumbai  | 1       |
| Chennai | 2       |

can create a false relationship where the model may assume:

```
Chennai > Mumbai > Delhi
```

or that the numbers have some mathematical importance.

One-Hot Encoding removes this problem because every category is treated equally.

---

## Where Is One-Hot Encoding Used?

One-Hot Encoding is commonly used for:

* **Machine learning classification problems**
* **Regression models with categorical features**
* **Natural language processing (NLP)**
* **Customer data analysis**
* **Recommendation systems**
* **Business and marketing datasets**

Examples of categorical features where it is useful:

* Gender: Male, Female
* Country: India, USA, UK
* Product Category: Mobile, Laptop, Tablet
* Department: HR, IT, Finance

---

## How Does One-Hot Encoding Help in Training Models?

Machine learning models such as:

* Linear Regression
* Logistic Regression
* Support Vector Machines
* Neural Networks
* K-Nearest Neighbors

require numerical input.

One-Hot Encoding helps by:

1. **Converting text data into numerical data**

   * Models can process the data mathematically.

2. **Avoiding incorrect relationships between categories**

   * Unlike label encoding, no category is considered greater or smaller.

3. **Improving model accuracy**

   * The model learns the actual effect of each category instead of learning false numerical patterns.

4. **Making features independent**

   * Each category gets its own feature column.

---

## One-Hot Encoding vs Label Encoding

| Feature           | Label Encoding        | One-Hot Encoding        |
| ----------------- | --------------------- | ----------------------- |
| Output            | Single numeric column | Multiple binary columns |
| Suitable for      | Ordered categories    | Unordered categories    |
| Example           | Low, Medium, High     | Red, Blue, Green        |
| Creates ranking?  | Yes                   | No                      |
| Number of columns | Less                  | More                    |

---

## Limitation of One-Hot Encoding

One-Hot Encoding can create a large number of columns when a feature has many unique categories. This is called the **curse of dimensionality**.

Example:

A feature like:

```
Country = 200 different countries
```

will create 200 new columns.

For high-cardinality features, techniques like **target encoding**, **frequency encoding**, or **feature hashing** may be more efficient.
