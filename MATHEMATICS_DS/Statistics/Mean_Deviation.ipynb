```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Sec_A = np.array([75, 65, 73, 68, 72, 76])
Sec_B = np.array([91, 49, 97, 93, 54, 45])
no = np.array([1, 2, 3, 4, 5, 6])
```

Mean

```python
mean_A = np.mean(Sec_A)
mean_B = np.mean(Sec_B)
mean_A , mean_B
```

Mean Absolute Deviation 

Formula: Mean Deviation = **∑i = 1^n |xi - x̄| / n**

where, 

∑ – Summation

xi – Observations

μ(x̄)  – Mean

n – Number of observations

```python
MAD_A = np.sum(np.abs(Sec_A - mean_A))/len(Sec_A)
MAD_B = np.sum(np.abs(Sec_B - mean_B))/len(Sec_B)
MAD_A, MAD_B
```

Mean Deviation Visualisation 

```python

plt.scatter(Sec_A, no, label="Sec_A")
plt.scatter(Sec_B, no, label= "Sec_B", c="RED")
plt.plot([71.5, 71.5, 71.5, 71.5, 71.5, 71.5], no, c= "blue", label= "Mean")
plt.legend()
plt.show()
```
