```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
set = {
    'Name' : ["Ashish", "Sneha", "Deepak", "Sakshi", "Kiran", "Rakesh", "Priya", "Bhawna", "Puneet", "Anil"],
    'Age'  : [23, 20, 21, 19, 21, 45, 23, 17, 34, 39]
}
df = pd.DataFrame(set)
df.head()
```

```python
mn = np.mean(df["Age"])
print(mn)
```

                                                       VISUALISATION                                                                       

```python
sns.histplot(x = "Age", data=df, bins= [i for i in range(0, 50 , 10)])
plt.plot([mn for i in range(0, 10)], [i for i in range(0, 10)], c = "RED")
plt.show()
```
