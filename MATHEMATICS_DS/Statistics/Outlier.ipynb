```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_excel(r"C:\Users\praja\OneDrive\Codes_Project\Data_Science\MATHEMATICS_DS\Statistics\100 People.xlsx")
dataset.head()
```

```python
dataset.info()
```

```python
sns.histplot(x="Value", data=dataset)
plt.show()
```

```python
# OUTLIERS REMOVE BY IQR METHOD 
Q1 = dataset['Value'].quantile(0.25)
Q3 = dataset['Value'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Remove outliers
clean_data = dataset[(dataset['Value'] >= lower) & (dataset['Value'] <= upper)]

print("Cleaned Data:")
print(clean_data)
```
