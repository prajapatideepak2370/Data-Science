```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
set = {
    'Name' : ["Ashish", "Sneha", "Deepak", "Sakshi", "Kiran", "Rakesh", "Priya", "Bhawna", "Puneet", "Anil"],
    'Age'  : [23, 20, 21, 19, 21, 45, 23, 17, 34, 39],
    'Salary' : [4500, 5000, 8000, 6000, 8500, 12000, 3400, 6500, 7800, 9500],
    "City" : ["Goa", "Delhi", "Goa", "Punjab", "Punjab", "Dubai", "Mumbai", "Punjab", "Delhi", "Punjab"]
}
df = pd.DataFrame(set)
df.head()
```

```python
md = df["City"].mode()[0]
print(md)
```

```python
df["City"].value_counts()
```

```python
order = sorted(df["City"].unique())
sns.countplot(x="City", data=df, order=order)

x_loc = order.index(md)
plt.axvline(x=x_loc, color="red", linestyle="--", label=f"Mode: {md}")

plt.legend()
plt.show()
```
