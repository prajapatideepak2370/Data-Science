```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = np.random.normal(0,100,100)
```

```python
#POSITIVE SKEWED VALUE 
df = pd.DataFrame({"x":data})
df["x"].skew()
```

```python
sns.histplot(x= "x", data=df, bins=[0,20,40,60,80,90,110,120,130,140,150,170,180])
plt.show()
```
