```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
set = {
    'Name' : ["Ashish", "Sneha", "Deepak", "Sakshi", "Kiran", "Rakesh", "Priya", "Bhawna", "Puneet", "Anil"],
    'Age'  : [23, 20, 21, 19, 21, 45, 23, 17, 34, 39],
    'Salary' : [4500, 5000, 8000, 6000, 8500, 12000, 3400, 6500, 7800, 9500]
}
df = pd.DataFrame(set)
df.head()
```

```python
np.sort(df["Salary"])
```

```python
np.count_nonzero(df["Salary"])
```

```python
mn = np.median(df["Salary"])
print(mn)
```

```python
nm = np.mean(df["Salary"])
print(nm)
```

```python
min_salary = df['Salary'].min()
sns.histplot(x = "Salary", data=df, bins=[i for i in range(2000, 14000, 1000)], kde= False)
plt.plot([mn for i in range(0, 5, 1)], [i for i in range(0, 5, 1)], c = "RED")
plt.plot([nm for i in range(0, 5, 1)], [i for i in range(0, 5, 1)], c = "Green")
plt.show()
```
